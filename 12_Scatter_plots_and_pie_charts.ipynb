# **`matplotlib` scatter plots**
---

> A `scatter` plot is a chart type that is normally used to observe and visually display the relationship between variables

### `plt.scatter()`  
> Some of the basic parameters: `plt.scatter(x, y, s=None, c=None, marker=None)`
+ `x, y` - The data positions
+ `c` - The marker colors
+ `marker` : `default 'o': ` - The marker style

```python
# Imports
import matplotlib.pyplot as plt
# Select stylesheet
plt.style.use("ggplot")

# The lists represent female and male body weights for various heights
heights = [137,140,142,145,147,150,152,155,157,160]
f_weights = [28.5,30.8,32.6,34.9,36.4,39,40.8,43.1,44.9,47.2]
m_weights = [34.9,38.1,33.5,35.8,46.7, 42.8,43.1,45.8,50.8,58.9]
```

```python
# Generating the plot
plt.figure(figsize=(6,5))
plt.title("Female / Male Weights against Height", pad=10)
plt.scatter(heights, f_weights, marker="o", color="#7c0902", label="Female")
plt.scatter(heights, m_weights, marker="X", color="#0070bb", label="Male")
plt.xlabel("Height (cm)", labelpad=8)
plt.ylabel("Weight (kg)", labelpad=8)
plt.legend(shadow=True, facecolor="#ffffff", ncols=3)
```

# **`matplotlib` pie charts**
---

> A `pie` chart is a type of graph representing data in a circular form, with each slice of the circle representing a fraction or proportionate part of the whole.

### `plt.pie()`
> Some of the basic parameters: `plt.pie(x, explode=None, labels=None, colors=None, shadow=False, autopct=None)`
+ `x` : `1D array-like` - Wedge sizes
+ `explode` : `array-like, default: None` - is a `len(x)` array which specifies the fraction of the radius with which to offset each wedge
+ `labels` : `list, default: None` - A sequence of strings providing the labels for each wedge
+ `colors` : `array-like of color, default: None` - A sequence of colors through which the pie chart will cycle
+ `shadow` : `bool, default: False` - If bool, whether to draw a shadow beneath the pie
+ `autopct` : `None or str or callable, default: None`- a string or function used to label the wedges with their numeric value. The label will be placed inside the wedge.

```python
# Lists to use:
labels = ["Turkey", "Potatoes", "Pumpkin Pie", "Stuffing"]
prices = [25.99,3.24,9.50, 6.99]
```

```python
# Generating the pie plot
plt.figure(figsize=(5,5))
plt.title("Thanksgiving Price Breakdown")
plt.pie(prices, labels=labels, explode=[0, 0, 0.1, 0], shadow=True, autopct="%1.1f%%")
plt.show()
```
