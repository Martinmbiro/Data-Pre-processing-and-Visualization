# **Adding and Removing Columns & Rows**
---
+ The `DataFrame.drop()` method is used for dropping both rows and columns
+ The `DataFrame.insert()` method is used for inserting columns
+ `df["new_col_name"] = ` is also used to add a new column to the DataFrame

### Parsing the Datasets into a DataFrame

```python
import pandas as pd

btc = pd.read_csv("/content/coin_Bitcoin.csv", index_col="Date")
houses = pd.read_csv("/content/kc_house_data.csv")
titanic = pd.read_csv("/content/titanic.csv")
countries = pd.read_csv("/content/world-happiness-report-2021.csv", index_col="Country name")
netflix = pd.read_csv("/content/netflix_titles.csv", delimiter="|", index_col=0)
```

### Dropping Columns
> `DataFrame.drop(columns = [list with column names], inplace = True/False)`
+ Returns a new DataFrame without altering the original, unless `inplace=True` is specified to actually modify the original `DataFrame`

```python
# Check out the bitcoin data set, that has two "useless" rows: Name & Symbol
btc.head(3)
```

```python
# The DataFrame also has a RangeIndex, hence the SNo column is also not needed
type(btc.index)
```

```python
# Dropping the Name, SNo, and Symbol columns
btc.drop(columns=["SNo", "Symbol", "Name"], inplace=True)
```

```python
btc.head()
```

### Dropping Rows

### Adding static columns
> A column that has the same value for every single row throughout the DataFrame

#### Using `df["new_col_name"] = `
> The new column gets appended to the end of the DataFrame

```python
# To the titanic DataFrame, add a new column called "species" with the value "human"
titanic.head(2)
```

```python
# Adding the column
titanic["species"] = "human"
```

```python
# Check this out:
titanic.head(2)
```

#### Using `df.insert(location, column_name, value)`
> One can specify the position location to insert the column

```python
# In the houses DataFrame, add a column named "county" right afte the "id" column
houses.head(2)
```

```python
# Adding the column
houses.insert(1, "county", "King County")
```

```python
# Viewing the DataSet:
houses.head(2)
```

### Creating new "Dynamic Columns"
> Values in the column are going to be calculated for each row, hence dynamic

#### `titanic` DataFrame
+ In the `titanic` DataFrame, add a column, `num_relatives` that will contain the number of relatives every passenger had, aboard the ship  
+ This number will be calculated by summing the `sibsp` and `parch` columns

```python
'''Drop the "species" column first'''
titanic.drop(columns="species", inplace=True)

# Adding the column
titanic.insert(8, "num_relatives", titanic["parch"]+titanic["sibsp"])
```

```python
# Check it out
titanic.head(3)
```

```python
# Sort by num_relatives
titanic.sort_values("num_relatives", ascending=False)
```

```python
# Find the people with the top num_relatives, who also survived
titanic[titanic["survived"]==1].sort_values("num_relatives", ascending=False)
```

```python
# Find the passengers that were travelling alone and find their passenger class distribution
titanic[titanic["num_relatives"]==0]["pclass"].value_counts()
```

#### Working with the `houses` DataFrame
+ In the `houses` DataFrame, add a new column, `price_sqft` to represent the value of price per square-foot
+ This value is to be calculated by `price / sqft_living`

```python
# Drop the county column first
houses.drop(columns="county", inplace=True)

# Adding the column
houses["price_sqft"] = houses["price"]/houses["sqft_living"]
```

```python
# Sort by the most expensive houses per sqft
houses.sort_values("price_sqft", ascending=False)
```

```python
# Which zipcode, has the most houses with highest price per sqft, among the top 50 houses with highest price per sqft
houses.nlargest(50, "price_sqft")["zipcode"].value_counts().plot(kind="barh", xlabel="No. Houses", ylabel="Zip Codes")
```

#### Working with the `btc` DataFrame
+ In the `btc` DataFrame, add a new column, `delta` to represent difference between the Bitcoin's `High` and `Low` prices
+ This value is to be calculated by `High - Low`

```python
# Addin the column
btc["delta"] = btc["High"] - btc["Low"]
```

```python
# Plot the top 10 days with highest delta
btc.nlargest(10, "delta").sort_values("delta", ascending=True)["delta"].plot(kind="barh", xlabel="High-Low Price")
```
