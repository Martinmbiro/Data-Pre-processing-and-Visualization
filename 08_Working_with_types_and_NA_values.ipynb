# **Working with types, and `NA` values**
---

+ `Series.astype()` - Casts a `pandas` object to a specified `dtype`
+ `pd.to_numeric()` - Converts data supplied to a numeric type
+ `.dropna()` - Removes missing values
+ `.isna()` - Returns a Boolean sized object indicating is values are `NA`  
+ `.fillna()` - Used to replace `NA` values with some other values

### Parsing the Datasets into DataFrames

```python
import pandas as pd
titanic = pd.read_csv("/content/titanic.csv")
```

### `Series.astype(dtype)`  
> Where `dtype` = `dtype`, `str`, data type, `Series` or Mapping of column name -> data type
+ Can be used to cast entire column into a specified `dtype`

#### A little background

```python
"""
Have a look at the titanic DataFrame
    The age column is an object dtype, but would much rather be a float
"""
titanic.info()
```

```python
titanic["age"].value_counts()
"""
The age column contains the string "?" therefore the dtype cannot be a float
  To accommodate the str, pandas sets the type to object instead
"""
```

#### Cast the `age` column in `titanic` DataFrame to `float`

```python
# Replace the "?" with Python value None to avoid running into any issues
titanic["age"].replace(["?"], [None], inplace=True)
```

```python
# Check to see that all "?" have been replaced to Python value None
titanic["age"].value_counts(dropna=False)
```

```python
# Casting
titanic["age"].astype("float")
# Returns a Series with all the values in the age column cast to float
```

```python
# Override the original age column with the new Series
titanic["age"] = titanic["age"].astype("float")

# age should now be a float64 type
titanic.info()
```

```python
# Now finding the mean age: (rounded to 2 decimal places)
round(titanic["age"].mean(),2)
```

### The `category` dtype
> Works in columns that have a set of possible finite values, for instance the `sex` column in `titanic` which can only contain either of two values, `male` or `female`
+ Helps optimize space for large Datasets

```python
# Setting sex and embarked columns to category
titanic["sex"] = titanic["sex"].astype("category")
titanic["embarked"] = titanic["embarked"].astype("category")

# Glimpse the dataframe
titanic.info()
```

### Casting with `pd.to_numeric(arg, errors="raise"/"coerce")`
+ Where `arg` = `Series`, `list`, e.t.c
+ When set to `errors="coerce"`,  then invalid parsing will be set as `NaN`
+ Returns `Series` if `Series`, was passed as `arg`, otherwise `ndarray`.

```python
# Re-importing the titanic dataset again
titanic2 = pd.read_csv("/content/titanic.csv")

# The age column is an object due to the "?"
titanic2.info()
```

```python
# Lots of "?"
titanic2["age"].value_counts()
```

```python
# Castin the age column to numeric
# Where there was a "?", the value will be set to NaN
pd.to_numeric(titanic2["age"], errors="coerce")
```

```python
# Returns a Series
type(pd.to_numeric(titanic2["age"], errors="coerce"))
```

```python
# Overwrite the existing age column in titanic2
titanic2["age"] = pd.to_numeric(titanic2["age"], errors="coerce")

# the age column should be a float
titanic2.info()
```

### `DataFrame.isna()` / `Series.isna()`
+ Return a boolean same-sized object indicating if the values are `NA`. `NA` values, such as `None` or `numpy.NaN`, gets mapped to `True` values.
+ Everything else gets mapped to `False` values.
> Returns a `Series` if called on a `Series`, and a `DataFrame` if called on a `DataFrame`

```python
# Read the game_stats Dataset
stats = pd.read_csv("/content/game_stats.csv")
stats
```

```python
stats.isna()
#When called on the entire DataFrame, returns a Boolean DataFrame
```

```python
# When called on a particular column / Series, returns a Series
stats["league"].isna()
```

```python
# Find specific rows without a league
stats[stats["league"].isna()]
```

### `dropna()`
> Works for both `Series` and `DataFrame` to remove missing values.  
>>Parameters:
+ `axis="columns" / "index"` - Columns and rows respctively (`DataFrame` use case)
+ `how="any" / "all"` - If `any` of the values is `NA`, drop the row or column; if `all` of the values are `NA`, drop the row or column (`DataFrame` use case)
+ `implace = True / False`
+ `subset = column label or sequence of labels` - Labels along other axis to consider, e.g. when dropping rows these would be a list of columns to include (`DataFrame` use case)


#### `Series.dropna()`
> When called on a `Series`, returns a `Series` with `NA` entries dropped from it or `None` if inplace=True

```python
# Working with a Series
stats["assists"]
```

```python
stats["assists"].dropna()
# Returns a Series with NA values dropped
```

#### `DataFrame.dropna()`
> When called on a `DataFrame`, returns a `DataFrame` with
`NA` entries dropped from it or `None` if inplace=True

```python
# Drop rows with at least an NA value somewhere in the row
stats.dropna(axis="index")
```

```python
# Only drop rows where all values are NA (the last row will be dropped)
stats.dropna(axis="index", how="all")
```

```python
# Specify to drop rows with NA values for particular columns, for instance, league
stats.dropna(subset=["league", "points"])
```

### `fillna()`
> Used to replace / fill `NA / NAN` values with some other values
>>Parameters:
+ `value = scalar, dictionary, Series, or DataFrame` - Value to use to fill holes (e.g. 0), alternately a dict/Series/DataFrame of values specifying which value to use for each index (for a Series) or column (for a DataFrame).
+ `axis = "index"` for `Series`; `axis = "index"/"columns"` for `DataFrame`
+ `implace = True / False`

```python
stats
```

```python
# Fill all NA values in the entire stats DataFrame with 0
stats.fillna(0)
```

```python
# Fill the NA values in the league column with "ameteur"
stats["league"].fillna("amateur", inplace=True)
stats
```

```python
# Specify values to replace with when working with an entire DataFrame (using a dictionary)
stats.fillna({"points":0, "assists":"none"})
```

```python
# Filling an NA with a value from the same row but a different column
sales = pd.read_csv("/content/sales.csv")
sales
```

```python
sales["shipping_zip"].fillna(sales["billing_zip"], inplace=True)
sales
```

**Work with the `netflix_titles.csv` dataset to answer the following questions:**

##### Set the index to be the `show_id` (inplace)

```python
# Read the dataset and set index as show_id
netflix = pd.read_csv("/content/netflix_titles.csv", delimiter="|", index_col="show_id")

# Detele the unnamed clumn at position 0 inplace
netflix.drop(columns=netflix.columns[0], inplace=True)
```

```python
netflix.info()
```

##### Quite a few rows are missing a value for `country`. Find the rows with no country

```python
netflix[netflix["country"].isnull()]
```

##### Find the rows with no `director`, `cast`, AND `country` listed

```python
# Create a mask with the conditions:
mask_dcc = netflix["director"].isna() & netflix["cast"].isna() & netflix["country"].isna()

# Plug the mask into the dataframe:
netflix[mask_dcc]
```

##### Drop all rows in the dataframe that have at least one `NA` value (not inplace!)

```python
netflix.dropna(axis="index", how="any")
```

##### Drop all columns in the dataframe that contain at least one `NA` value, not inplace!

```python
netflix.dropna(axis="columns", how="any")
```

##### Drop all rows in the dataset that have a missing `director` or `cast` (not inplace!)

```python
netflix.dropna(subset=["director", "cast"])
```

##### Find the rows that are missing a `rating`  
> Replace the `na` ratings with `TV-MA` using `.loc[]`

```python
netflix.loc[netflix["rating"].isna(), ["rating"]] = "TV-MA"
```

##### Fill the missing `country` values with the most common (`mode`) `country` value, inplace!
> Do this using `fillna`, rather than `.loc[]`

```python
# Mode contry value
netflix["country"].mode()
```

```python
# Use a dictionary
netflix.fillna({"country":"United States"})
```
