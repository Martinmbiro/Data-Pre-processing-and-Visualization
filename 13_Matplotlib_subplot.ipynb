# **matplotlib `subplot`**
---

> With the `plt.subplot()` function one can draw multiple plots in one `figure`. It adds an `Axes` to the current `figure`.

> Most Common call signature (granted, there can be others):
>> `plt.subplot(nrows, ncols, index, **kwargs)`
+ `nrows` - no. of rows that the figure contains
+ `ncols` - no. of columns that the figure contains
+ `index` -  represents the index of the current plot, with index starting at `1`

> This method returns an `Axes`, i.e, one `subplot` in  a `figure`

### Tinkering Around

```python
# Importing libraries
import matplotlib.pyplot as plt
import numpy as np

# Specify the stylesheet to use
plt.style.use("ggplot")

# Create an array
nums = np.arange(5) # [0 1 2 3 4]
```

```python
# Generating 1 by 3 subplot
plt.figure(figsize=(10,4))
plt.suptitle("1 By 3 Subplot Sample", fontsize=15)

# Plot 1
plt.subplot(1,3,1)
plt.plot(nums)
plt.title("x", fontsize=10)

# Plot2
plt.subplot(1,3,2)
plt.plot(nums, nums**2)
plt.title("x^2", fontsize=10)


# Plot 3
plt.subplot(1,3,3)
plt.plot(nums, nums**3, )
plt.title("x^3", fontsize=10)

plt.show()
```

```python
# Generating 2 by 2 subplot
plt.figure(figsize=(8,8))
plt.suptitle("2 by 2 Subplot Sample", fontsize=15)

# Plot 1
plt.subplot(2,2,1)
plt.plot(nums)

# Plot 2
plt.subplot(2,2,2)
plt.plot(nums, nums**2)

# Plot 3
plt.subplot(2,2,3)
plt.plot(nums)

# Plot 4
plt.subplot(2,2,4)
plt.plot(nums, nums**3)

plt.show()
```

### Working with the Titanic dataset
> Plot the age distribution of passengers in the first, second and third class of the titanic on a `hist`
+ Ensure that the subplots share the same `yaxis`

```python
# Import pandas
import pandas as pd

# Parse the dataset into a Dataframe
titanic = pd.read_csv("/content/titanic.csv")
titanic.info()
```

```python
# The age column is an object, hence has to be converted to a numeric
titanic["age"] = pd.to_numeric(titanic["age"], errors="coerce")

# Create variables of ages in first, second and third class
first_class = titanic[titanic["pclass"]==1]["age"]
second_class = titanic[titanic["pclass"]==2]["age"]
third_class = titanic[titanic["pclass"]==3]["age"]
```

```python
# Generating the histogram
plt.figure(figsize=(10,4))
plt.suptitle("Titanic Age Distribution", fontsize=15)

# Plot 1
""" Save the first plot to a variable ax"""
ax = plt.subplot(1,3,1)
plt.hist(first_class, color="#f1c407")
plt.title("First Class", fontsize=10)
plt.xlim(0,90)

# Plot 2
plt.subplot(1,3,2, sharey=ax)
plt.hist(second_class, color="#3498db")
plt.title("Second Class", fontsize=10)
plt.xlim(0,90)

# Plot 3
plt.subplot(1,3,3, sharey=ax)
plt.hist(third_class, color="#e74c3c")
plt.title("Third Class", fontsize=10)
plt.xlim(0,90)

plt.show()
```

# **matplotlib `subplots`**
---

> The [`plt.subplots()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html) function is used to create a `figure` and an `Axes` (subplot) / set of `Axes`_s_ (subplots) in a single call.

> This method returns the following two items:
+ A [`figure`](https://matplotlib.org/stable/api/figure_api.html)  
+ An [`Axes`](https://matplotlib.org/stable/api/axes_api.html) or a _**set**_ of `Axes`  

> Common method call signature:  
>> `plt.subplots(nrows=1, ncols=1, sharex=False, sharey=False, **fig_kw)`
+ `nrows` - no. of rows that the figure contains
+ `ncols` - no. of columns that the figure contains
+ `sharex, sharey` -  Controls sharing of properties among x (sharex) or y (sharey) axes  
+ `**fig_kw` - All additional keyword arguments are passed to the `pyplot.figure` call.

#### Line plot with sin*

```python
import matplotlib.pyplot as plt
import numpy as np

# Specify the stylesheet to use
plt.style.use("ggplot")

# Create some toy data
x = np.linspace(0, 2 * np.pi, 400)
y1 = np.sin(x)
y2= np.sin(x**2)

# plt.subplots() returns a list of Axes since we've specified 2 columns
fig, axs = plt.subplots(1, 2, sharey=True, figsize=(7,3), layout='constrained')

# Making the plots
axs[0].plot(x, y1, color="#ab274f", linewidth=2)
axs[1].plot(x, y2, color="#003262", linewidth=2)

# Define the subplots titles
axs[0].set_title("Simple plot with sin(x)", fontsize=10)
axs[1].set_title("Simple plot with sin(x^2)", fontsize=10)

# Figure configuration
fig.suptitle("Stacked line subplots in one direction", fontsize=15)

plt.show()
```

#### Plotting categorical values in bar, scatter and line

```python
import matplotlib.pyplot as plt

# Specify stylesheet
plt.style.use("ggplot")

data = {'Apple': 10, 'Orange': 15, 'Lemon': 5, 'Lime': 20}
names = list(data.keys())
values = list(data.values())

# Call the method
fig, axs = plt.subplots(1, 3, figsize=(9.5,4), layout="constrained", sharey=True)

# Make the plots
axs[0].bar(names, values, color="#0047ab")
axs[0].set_title("Bar Plot", fontsize=10)

axs[1].plot(names, values, color="#de3163")
axs[1].set_title("Line Plot", fontsize=10)

axs[2].scatter(names, values, color="#7c0902")
axs[2].set_title("Scatter Plot", fontsize=10)

# Figure configuration
fig.suptitle("Plotting Categorical Data", fontsize=15)

plt.show()
```

#### Bar plot with various colors

```python
import matplotlib.pyplot as plt

# Specify stylesheet
plt.style.use("ggplot")

# The data
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

# The bar plot
fig, ax = plt.subplots(figsize=(5.5,5.5))
ax.bar(fruits, counts, color=bar_colors, label=bar_labels)

# Figure configurations
fig.legend(title="Fruits", ncols=4, fancybox=True, facecolor="#ffffff", loc="upper center")

plt.show()
```

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Specify the stylesheet
plt.style.use("ggplot")

# Parsing the dataset into a dataframe
sales = pd.read_csv("/content/kc_house_data.csv")

# Convert the date column into datetime
sales["date"] = pd.to_datetime(sales["date"])
```

```python
# Finding sales per month
sales["date"].dt.month.value_counts().sort_index()
```

```python
# Finding sales per day of week
sales["date"].dt.dayofweek.value_counts().sort_index()
```

```python
# The data:
monthly_sales = sales["date"].dt.month.value_counts().sort_index()
daily_sales = sales["date"].dt.dayofweek.value_counts().sort_index()
days = ["Mon", "Tue", "Wed", "Thur", "Fri", "Sat", "Sun"]
```

```python
# Putting it all together
fig, axs = plt.subplots(1, 2, figsize=(8,4), layout="constrained")

# The bar sub-plot
axs[0].bar(days, daily_sales, color="#722F37")
axs[0].set_ylabel("House Sales", fontsize=11)
axs[0].set_xlabel("Day of the Week", fontsize=10)
axs[0].set_title("Sales by Weekday", fontsize=11)

# Plot the line sub-plot
axs[1].plot(monthly_sales, color="#00009c", marker="*", markeredgecolor="#e60026", markerfacecolor="#e60026", linewidth=2)
axs[1].set_xlim(1,12)
axs[1].set_xticks(np.arange(1,13))
axs[1].set_xlabel("Month", fontsize=10)
axs[1].set_title("Sales by Month", fontsize=11)

# Figure configurations
fig.suptitle("King County House Sales", fontsize=15)

plt.show()
```
