# **`matplotlib` bar plots**
---

> A `bar` plot is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent

### `plt.bar()`  
> Some of the basic parameters: `plt.bar(x, height, width=0.8, bottom=None, align='center / edge')`
+ `x` - x coordinates of the bar
+ `height` - y coordicates of the bars (values of the bars)
+ `bottom` - The y coordinate(s) of the bottom side(s) of the bars
+ `align = center / edge`
>> `center` - Center the base on the x positions  
>> `edge`  - Align the left edges of the bars with the x positions


```python
# Import pyplot
import matplotlib.pyplot as plt

# Set the stylesheet for the plots
plt.style.use("ggplot")

#Lists representing seed data taken from a greenhouse:
plants = ['Spinach', 'Turnip', 'Rhubarb', 'Broccoli', 'Kale']
died = [10,25,5,30,21]
germinated = [74, 88, 56,69,59]
```

```python
# Plotting the seeds that did not germinate
plt.figure(figsize=(6,5))
plt.title("Seeds That DID NOT Germinate")
plt.xlabel("Seed Type", labelpad=8)
plt.ylabel("No. of Seeds", labelpad=5)
plt.bar(plants, died, color="#bf4f51")
```

```python
# Plotting the seeds that germinated
plt.figure(figsize=(6,5))
plt.title("Seeds That Germinated")
plt.xlabel("Seed Type", labelpad=8)
plt.ylabel("No. of Seeds", labelpad=5)
plt.bar(plants, germinated, color="#317873")
```

```python
# Having multiple plots on the same pair of axis
plt.figure(figsize=(6,5))
#plt.title("Seeds That Germinated")
plt.xlabel("Seed Type", labelpad=8)
plt.ylabel("No. of Seeds", labelpad=5)
plt.bar(plants, germinated, color="#317873")
plt.bar(plants, died, color="#bf4f51", align="edge", width=0.4)
```

```python
# Making a stacked bar plot
plt.figure(figsize=(6,5))
plt.title("Greenhouse Seeds Data")
plt.xlabel("Seed Type", labelpad=8)
plt.ylabel("No. of Seeds", labelpad=5)
plt.bar(plants, germinated, color="#317873", label="Germinated Seeds")
plt.bar(plants, died, color="#bf4f51", bottom=germinated, label="Failed Seeds")
plt.legend(shadow=True, facecolor="#ffffff")
```

# **`matplotlib` histograms**
---

> A `hist` is a graph used to represent the frequency distribution of a few data points of one variable

### `plt.hist()`

> Some of the basic parameters: `plt.hist(x, bins=None, range=None)`
+ `x` - Input values
+ `bins` : `default = 10` - defines the number of equal-width bins in the range
+ `range` : `tuple or None, default: None` - The lower and upper range of the bins. Lower and upper outliers are ignored

```python
# Plotting the age distribution of the passengers aboard Passenger Class 3 in Titanic
import pandas as pd

# parse the Dataset into a DataFrame
titanic = pd.read_csv("/content/titanic.csv")

# Ensure all values in the age column are numeric in nature, any exceptions are set to NaN
titanic["age"] = pd.to_numeric(titanic["age"], errors="coerce")
```

```python
# Now plot the graph
plt.figure(figsize=(6,5))
plt.hist(titanic[titanic["pclass"]==3]["age"], color="#007ba7")
plt.title("Age Distribution in P.Class-3 of the Titanic")
plt.xlabel("Passengers' Ages", labelpad=8)
plt.ylabel("Frequency", labelpad=5)
plt.xlim(0,80)
```
