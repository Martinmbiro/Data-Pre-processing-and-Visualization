# **`matplotlib` Intro**
---

> Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations.  


### `plt.plot()`
+ Plot y versus x as lines and / or markers
> `plot(x, y)` -  Plots x and y using default line style and color

+ If only one list of values is provided, the values are treated as the `yaxis` values, and values on the `xaxis` are set from `1...`

>> [Documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```python
import matplotlib.pyplot as plt

# If matplotlib.pyplot() is called, a blank graph is returned as such:
plt.plot()
```

```python
"""
Providing only one list treats those values as yaxis values
  xaxis values will begin from 1...
"""
plt.plot([2,4,2,6,8,4])
```

```python
# Let's provide two lists for x and y, shall we?
salaries = [55000, 65000, 72000, 90000, 115000, 150000]
ages=[20, 25, 30, 32, 40, 45]

# Order of axes is: plt.plot(x, y)
plt.plot(ages, salaries)
```

### Do we really need `plt.show()`_?_  

+ When working in a JupyterNotebook environment or something similar to it, `plt.plot()` will automatically be rendered, hence `plt.show()` is not necessary
+ However, when working in a different environment, `plt.show()' has to be called, to _display all open figures_

```python
plt.plot(ages, salaries)
plt.show()
```

### Matplotlib `figures`  

> The `figure` is like a canvas on which the plot is presented
+ `matplotlib.pyplot.figure` creates a new figure, or activate an existing figure.
+ `plt.figure(figsize = touple)` will change the dimenion of the figure `(width, height)` respectively. Units are in inches
>> Example: `plt.figure(figsize=(3,5))`

```python
# Create a numpy array so that it's easier to generate arrays
import numpy as np
nums = np.arange(5)

nums
```

```python
"""
Watch the behaviour of the plot
  All three plots will appear on the same FIGURE
"""
plt.plot(nums, nums)
plt.plot(nums, nums**2)
plt.plot(nums, nums**3)
```

```python
# Now calling plt.figure() will separate the plots after the method call

# Figure 1
plt.figure(figsize=(3,3))
plt.plot(nums, nums)

# Figure 2
plt.figure(figsize=(3,3))
plt.plot(nums, nums**2)
plt.plot(nums, nums**3)
```

```python
# Figure 1
plt.figure(figsize=(3,3))
plt.plot(nums, nums)

# Figure 2
plt.figure(figsize=(3,3))
plt.plot(nums, nums**2)

# Figure 3
plt.figure(figsize=(3,3))
plt.plot(nums, nums**3)
```

+ To use any one of these stylesheets, run `plt.style.use("style_sheet_name")`

 ### Line Styles, Colors, Widths, and More!  
 + Line properties can be modified
 > `line2D properties` can be found in the documentation [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D)

```python
# Setting line width and color

plt.figure(figsize=(5,5))
plt.plot(nums, nums, color="#be4d25", linewidth=3)
plt.plot(nums, nums**2, color="#2596be")
plt.plot(nums, nums**3, color="#6c25be", linewidth=2)
```

```python
# Line style
plt.figure()
plt.plot(nums, nums, color="#be4d25", linestyle="--")
plt.plot(nums, nums**2, color="#2596be", linestyle="-.")
plt.plot(nums, nums**3, color="#6c25be")
```

```python
# Line marker
plt.figure(figsize=(5,5))
plt.plot(nums, nums, color="#be4d25", marker="*", markerfacecolor="#041014", markeredgecolor="#041014")
```

### Plot labels and titles  
+ The plots can have titles assigned to them, as well as `xlabel` and `ylabel`
> More from the documentation here [xlabel](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html) and here [ylabel](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylabel.html)

```python
# Assigning a title to the plot
plt.figure(figsize=(5,5))
plt.plot(ages, salaries)
plt.title("Avg Salaries vs Employee Ages")
```

```python
# Labelling the x and y axes
plt.figure(figsize=(5,5))
plt.plot(ages, salaries)
plt.title("Avg Salaries vs Employee Ages", color="olive")
plt.xlabel("Employee Ages")
plt.ylabel("Average Salary", labelpad=10)
```

### Changing x and y ticks
+ Ticks are the values used to show specific points on the coordinate axis. Ticks can be numbers or strings  
+ A `list` of tick locations is passed to `plt.xticks()` and `plt.yticks()`  
+ Limits for the x and y axes can be set as well using `plt.xlim(bottom, top)` and `plt.ylim(bottom, top)`

> More information can be found here [plt.xticks()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xticks.html) and here [plt.yticks()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.yticks.html)  
> Also, here [plt.xlim()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlim.html) and here[plt.ylim()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylim.html)

```python
# Adding xticks
plt.figure(figsize=(5,5))
plt.plot(ages, salaries)
plt.title("Avg Salaries vs Employee Ages")
plt.xticks([20,30,40,50,60])
plt.yticks([60000,80000,100000,120000,140000])
```

```python
# Adding tick labels
plt.figure(figsize=(5,5))
plt.title("Avg Salaries vs Employee Ages")
plt.plot(ages, salaries)
plt.xticks([20,30,40,50,60])
plt.yticks([60000,80000,100000,120000,140000], labels=["60K", "80K", "100K", "120K", "140K"])
```

```python
# Specifying limits for the axes
plt.figure(figsize=(5,5))
plt.title("Avg Salaries vs Employee Ages")
plt.plot(ages, salaries)
plt.xticks([20,30,40,50,60])
plt.yticks([60000,80000,100000,120000,140000], labels=["60K", "80K", "100K", "120K", "140K"])
plt.xlim(20,60)
plt.ylim(55000)
```

### Adding legends to plots  
+ The legend is a key that labels various plot elements

```python
plt.figure(figsize=(4.5,4.5))
plt.title("Nums vs Result")
plt.plot(nums, color="#be4d25", label="num^1", linewidth=1.5)
plt.plot(nums**2, color="#2596be", label="num^2", linewidth=1.5)
plt.plot(nums**3, color="#6c25be", label="num^3", linewidth=1.5)
plt.xlabel("Nums")
plt.ylabel("Result")
plt.xticks([0,1,2,3,4,5,6])
plt.ylim(0,70)
plt.xlim(0,5)
plt.legend(shadow=True, facecolor="#ffffff")
```
